```python
import sklearn.decomposition as dc
import numpy as np
import matplotlib.pyplot as plt
#%%
t = np.linspace(0,100,100)
x1 = np.sin(2*np.pi*t*0.01)
x2 = np.sin(2*np.pi*t*0.02)
x3 = np.sin(2*np.pi*t*0.03)
x4 = np.sin(2*np.pi*t*0.04)
x = np.c_[x1,x2,x3,x4]

for i in range(np.size(x,1)):
    plt.subplot(4,4,i+1)
    plt.plot(x[:,i])
    plt.title('original')

matriz_ruido = np.random.randn(np.size(x,1),np.size(x,1))
x_ruido = np.dot(x,matriz_ruido)

for i in range(np.size(x,1)):
    plt.subplot(4,4,i+5)
    plt.plot(x_ruido[:,i])
    plt.title('ruidosa')
    
pca = dc.PCA(np.size(x,1),whiten = True, random_state = 42)

x_pca = pca.fit_transform(x_ruido)

for i in range(np.size(x,1)):
    plt.subplot(4,4,i+9)
    plt.plot(x_pca[:,i])
    plt.title('PCA')

matriz_in_out = np.linalg.inv(pca.components_)

x_conv = np.dot(x_ruido,matriz_in_out)

for i in range(np.size(x,1)):
    plt.subplot(4,4,i+13)
    plt.plot(x_conv[:,i])
    plt.title('Recons')
```
